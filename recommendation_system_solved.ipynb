```python
import re
def text_cleaning(text):
    text = re.sub(r'&quot;', '', text)
    text = re.sub(r'.hack//', '', text)
    text = re.sub(r'&#039;', '', text)
    text = re.sub(r'A&#039;s', '', text)
    text = re.sub(r'I&#039;', 'I\'', text)
    text = re.sub(r'&amp;', 'and', text)
    
    return text

data["name"] = data["name"].apply(text_cleaning)
```

```python
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors

data_pivot = data.pivot_table(index="name",columns="anime_id",values="rating").fillna(0)
data_matrix = csr_matrix(data_pivot.values)
model_knn = NearestNeighbors(metric = "cosine", algorithm = "brute")
model_knn.fit(data_matrix)

query_no = np.random.choice(data_pivot.shape[0]) # random anime title and finding recommendation
print(f"We will find recommendation for {query_no} no anime which is {data_pivot.index[query_no]}.")
distances, indices = model_knn.kneighbors(data_pivot.iloc[query_no,:].values.reshape(1, -1), n_neighbors = 6)
```

```python
no = []
name = []
distance = []
rating = []

for i in range(0, len(distances.flatten())):
    if i == 0:
        print(f"Recommendations for {data_pivot.index[query_no]} viewers :\n")
    else:
        #  print(f"{i}: {data_pivot.index[indices.flatten()[i]]} , with distance of {distances.flatten()[i]}")        
        no.append(i)
        name.append(data_pivot.index[indices.flatten()[i]])
        distance.append(distances.flatten()[i])
        rating.append(*df[df["name"]==data_pivot.index[indices.flatten()[i]]]["rating"].values)

dic = {"No" : no, "Anime Name" : name, "Rating" : rating}
recommendation = pd.DataFrame(data = dic)
recommendation.set_index("No", inplace = True)
recommendation.style.set_properties(**{"background-color": "#2a9d8f","color":"white","border": "1.5px  solid black"})
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfv = TfidfVectorizer(min_df=3, max_features=None, strip_accents="unicode", analyzer="word",
                      token_pattern=r"\w{1,}", ngram_range=(1, 3), stop_words = "english")

rec_data = df.copy()
rec_data.drop_duplicates(subset ="name", keep = "first", inplace = True)
rec_data.reset_index(drop = True, inplace = True)
genres = rec_data["genre"].str.split(", | , | ,").astype(str)
tfv_matrix = tfv.fit_transform(genres)
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

sig = sigmoid_kernel(tfv_matrix, tfv_matrix)      # Computing sigmoid kernel

rec_indices = pd.Series(rec_data.index, index = rec_data["name"]).drop_duplicates()


# Recommendation Function
def give_recommendation(title, sig = sig):
    
    idx = rec_indices[title] # Getting index corresponding to original_title

    sig_score = list(enumerate(sig[idx]))  # Getting pairwsie similarity scores 
    sig_score = sorted(sig_score, key=lambda x: x[1], reverse=True)
    sig_score = sig_score[1:11]
    anime_indices = [i[0] for i in sig_score]
     
    # Top 10 most similar movies
    rec_dic = {"No" : range(1,11), 
               "Anime Name" : df["name"].iloc[anime_indices].values,
               "Rating" : df["rating"].iloc[anime_indices].values}
    dataframe = pd.DataFrame(data = rec_dic)
    dataframe.set_index("No", inplace = True)
    
    print(f"Recommendations for {title} viewers :\n")
    
    return dataframe.style.set_properties(**{"background-color": "#2a9d8f","color":"white","border": "1.5px  solid black"})
```

What is Collaborative Filtering?
Collaborative Filtering is a technique used in recommendation systems to predict the preferences or interests of a user by collecting preferences from many users. The basic idea is that if users agree on past preferences, they will agree on future preferences.

How Does Collaborative Filtering Work?
Data Collection: Gather data on user interactions with items (e.g., ratings, purchases, clicks).
Similarity Calculation: Calculate the similarity between users or items based on their interaction data.
Prediction: Use the similarity information to predict a user's preference for an item they haven't interacted with.
Recommendation: Recommend items to users based on the predicted preferences.
Types of Collaborative Filtering
There are two main types: User-Based Collaborative Filtering and Item-Based Collaborative Filtering.

User-Based Collaborative Filtering
How it Works:

Find Similar Users: Identify users who have similar preferences or behaviors to the target user.
Make Predictions: Recommend items that similar users have liked or interacted with but the target user hasn't yet.
Example:

If User A and User B both liked Movies X and Y, and User A also liked Movie Z, then User B might be recommended Movie Z.
Advantages:

Personal and intuitive: Recommendations are based on users with similar tastes.
Disadvantages:

Scalability: Can be computationally expensive with a large number of users.
Cold Start: Struggles with new users who have no interaction history.
Item-Based Collaborative Filtering
How it Works:

Find Similar Items: Identify items that are similar based on user interactions.
Make Predictions: Recommend items similar to those the user has interacted with or liked.
Example:

If many users who liked Movie X also liked Movie Y, then Movie Y might be recommended to users who liked Movie X.
Advantages:

Scalability: Generally more scalable than user-based methods.
Stable: Recommendations are based on item similarity, which doesn’t change as frequently as user preferences.
Disadvantages:

Less Personal: Recommendations may feel less personalized as they are based on item similarities rather than individual user tastes.
Key Differences
Focus:

User-Based: Focuses on the similarity between users.
Item-Based: Focuses on the similarity between items.
Scalability:

User-Based: Less scalable due to the need to compare all users.
Item-Based: More scalable as item similarities tend to be more stable and fewer in number compared to users.
Cold Start Problem:

User-Based: Struggles with new users.
Item-Based: Struggles with new items.
