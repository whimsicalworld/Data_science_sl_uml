```python
# Impute missing values
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)
titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0], inplace=True)
titanic_df['Cabin'].fillna('Unknown', inplace=True)

missing_values_after_imputation = titanic_df.isnull().sum()
missing_values_after_imputation
```

```python
# Encode categorical variables
titanic_df = pd.get_dummies(titanic_df, columns=['Sex', 'Embarked', 'Pclass', 'Cabin', 'Ticket'], drop_first=True)
```

```python
from sklearn.model_selection import train_test_split

# Drop the 'Survived' and 'Name' columns if 'Name' is not needed
X = titanic_df.drop(columns=['Survived', 'Name'])
y = titanic_df['Survived']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

```

```python
import lightgbm as lgb
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Initialize LightGBM and XGBoost classifiers
lgb_clf = lgb.LGBMClassifier(random_state=42)
xgb_clf = XGBClassifier(n_estimators=90,max_depth=None,learning_rate=0.1)

# Train the models
lgb_clf.fit(X_train, y_train)
xgb_clf.fit(X_train, y_train)

# Predict on the test set
y_pred_lgb = lgb_clf.predict(X_test)
y_pred_xgb = xgb_clf.predict(X_test)

# Evaluate the models
def evaluate_model(y_test, y_pred):
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    return accuracy, precision, recall, f1

lgb_metrics = evaluate_model(y_test, y_pred_lgb)
xgb_metrics = evaluate_model(y_test, y_pred_xgb)

lgb_metrics, xgb_metrics
```

```python
# Compare performance metrics
metrics_df = pd.DataFrame({
    'Metric': ['Accuracy', 'Precision', 'Recall', 'F1-Score'],
    'LightGBM': lgb_metrics,
    'XGBoost': xgb_metrics
})

metrics_df
```

```python
# Visualize the performance comparison
metrics_df.plot(x='Metric', y=['LightGBM', 'XGBoost'], kind='bar', figsize=(10, 6))
plt.title('Performance Comparison between LightGBM and XGBoost')
plt.ylabel('Score')
plt.show()
```

The comparative analysis between LightGBM and XGBoost on the Titanic dataset reveals that:


Both models perform well in terms of accuracy, precision, recall, and F1-score.

Minor differences in metrics suggest that either algorithm could be preferred based on specific needs and further tuning.

LightGBM is generally faster in training due to its design but XGBoost might offer slightly better interpretability and tuning flexibility.


Further hyperparameter tuning and cross-validation can be performed to optimize both models and identify the best model for practical deployment.
